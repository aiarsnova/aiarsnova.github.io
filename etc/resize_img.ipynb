# Resize by height

```python
from PIL import Image
import matplotlib.pyplot as plt

def resize_image_with_aspect_ratio(image_path, new_height, output_path):
    # 이미지 열기
    image = Image.open(image_path)
    
    # 원본 이미지 크기 가져오기
    original_width, original_height = image.size
    
    # 새로운 너비 계산
    aspect_ratio = original_width / original_height
    new_width = int(new_height * aspect_ratio)
    
    # 이미지 리사이즈
    resized_image = image.resize((new_width, new_height))
    print(image.size)
    print(resized_image.size)
    
    # 리사이즈된 이미지 저장
    # resized_image.save(output_path)
    
    # 리사이즈된 이미지 표시
    plt.imshow(image)
    plt.axis('off')  # 축 숨기기
    plt.show()
    plt.imshow(resized_image)
    plt.axis('off')  # 축 숨기기
    plt.show()
    
    print("이미지 리사이즈 완료!")

# 사용 예시
resize_image_with_aspect_ratio('/nas/users/dahye/2024_dahye_proj/aiarsnova_github/aiarsnova.github.io/images/memorial1.png', 400, 'path_to_save_resized_image.jpg')
```
